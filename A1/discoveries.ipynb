```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/15840/Desktop/2025Fall/05619/A1/fatal_selected.csv")
```

```python
age_counts = df["age"].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,6), dpi=130)
age_counts.plot(kind="bar")
plt.title("Distribution of Deaths by Exact Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
out_path = "/Users/15840/Desktop/2025Fall/05619/A1/pic/q1_age_distribution_exact.png"
plt.savefig(out_path)
plt.show()
```

```python
bins = list(range(0, 90, 10)) + [200]
labels = [f"{bins[i]}-{bins[i+1]-1}" if bins[i+1] != 200 else f"{bins[i]}+" for i in range(len(bins)-1)]
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels, right=False, include_lowest=True)

age_counts_group = df["age_group"].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,5), dpi=130)
age_counts_group.plot(kind="bar")
plt.title("Distribution of Deaths by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.tight_layout()
```

```python
out_path2 = "/Users/15840/Desktop/2025Fall/05619/A1/pic/q1_age_distribution_grouped.png"
plt.savefig(out_path2)
plt.show()
```

```python
zip_counts = df["incident_zip"].value_counts().head(10)
```

```python
plt.figure(figsize=(8,5), dpi=130)
zip_counts.plot(kind="bar")
plt.title("Top 10 ZIP Codes by Number of Deaths")
plt.xlabel("ZIP Code")
plt.ylabel("Count")
plt.tight_layout()
```

```python
df["death_date_and_time"] = pd.to_datetime(df["death_date_and_time"], errors="coerce")
df["year_from_date"] = df["death_date_and_time"].dt.year
```

```python
year_counts_date = df.groupby("year_from_date").size().sort_index()

plt.figure(figsize=(9,5), dpi=130)
plt.plot(year_counts_date.index, year_counts_date.values, marker="o")
plt.title("Deaths Over Time (by death_date_and_time)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
out_path_date = "/Users/15840/Desktop/2025Fall/05619/A1/pic/q3_time_trend_by_date.png"
plt.savefig(out_path_date)
plt.show()
```

```python
year_counts_case = df.groupby("case_year").size().sort_index()

plt.figure(figsize=(9,5), dpi=130)
plt.plot(year_counts_case.index, year_counts_case.values, marker="o")
plt.title("Deaths Over Time (by case_year)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
out_path_case = "/Users/15840/Desktop/2025Fall/05619/A1/pic/q3_time_trend_by_caseyear.png"
plt.savefig(out_path_case)
plt.show()
```
